## Question 2: Cleaned review with stems

```python
# Contraction expansion
!pip install contractions
import contractions
```

```python
# Define contraction funtion
def contraction_expansion(text):
  expanded_word = []    
  for word in text.split():
  # using contractions.fix to expand
    expanded_word.append(contractions.fix(word))   
   
  return  ' '.join(expanded_word)
```

```python
# Test to see if there was any empty reviews in the dataset
df['Review'].isnull().values.any()
```

```python
# There is no empty reviews, so we start to work on contraction 
df['Review_expanded'] = df['Review'].apply(contraction_expansion)
df.iloc[5:9,:]
```

```python
# tokenize each review in the dataframe
from nltk.tokenize import wordpunct_tokenize
%time df['Review_Token'] = df['Review_expanded'].apply(wordpunct_tokenize)
df.head(3)
```

```python
# Normalization with stemming
import nltk
from nltk.stem.snowball import SnowballStemmer

snow_stemmer = SnowballStemmer(language='english')  # set the language the parameter
  
#stem each word in the token list of each review
def snowball_Stemmer(token_col):
  """ stem each word in the specfical col """
  snowball_words = []
  for w in token_col:
    x = snow_stemmer.stem(w)
    snowball_words.append(x)
  return snowball_words
```

```python
# apply snowballstemmer
%time df['Review_Stem'] = df['Review_Token'].apply(snowball_Stemmer)
df.head(3)
```

```python
# Define preprocessing function
from nltk.corpus import stopwords
nltk.download('stopwords')
from string import punctuation

def preprocessing(tokens):
  tokens = [token.lower() for token in tokens] # lowercasing
  
  tokens = [token for token in tokens if not token.isdigit()] # remove digit

  tokens = [token for token in tokens if token not in punctuation] # remove punctuations

  mystopwords = set(stopwords.words("english")) 
  tokens = [token for token in tokens if token not in mystopwords] # remove stopwords
  
  tokens = [word for word in tokens if len(word)>=3] # remove words with one or two characters

  return tokens
```

```python
# Apply text preprocessing
df['Review_Stem_Final'] = df['Review_Stem'].apply(preprocessing) 
df.head(3)
```

```python
# find top 500 stems based on the count frequency
from nltk import FreqDist
# use FreqDis() to find the frequency disctibution for each token
%time fdist_reviews= FreqDist([token for sublist in df['Review_Stem_Final'].values for token in sublist])
print('The total number of tokens:',fdist_reviews.N())
print('The number of unique tokens:', len(fdist_reviews))
```

```python
# test on the a case
fdist_reviews['pizza']
```

```python
# let's sort the items based on the frequency count
fdist_reviews_list = sorted(fdist_reviews.items(), key=lambda x:x[1], reverse= True)
print(type(fdist_reviews_list)) # this is a list of tuples
print('The first item in the list:', fdist_reviews_list[0])
```

```python
# see the top 10 tokens and its count
for x in fdist_reviews_list[:10]:
    print (x[0], x[1])
```

```python
# create a list of top 500 tokens
token_list = [token[0] for token in fdist_reviews_list[:500]]
print(len(token_list))
```

```python
# define function to filter token
def remove500(tokens):
  tokens = [token for token in tokens if token in token_list] # keep the token in the token_list
  return tokens
```

```python
# Apply text filtering
df['Cleaned_Review_Stems'] = df['Review_Stem_Final'].apply(remove500) 
df.head(3)
```

## Question 5: Annotation with NER tags

```python
# create NER taggs using spaCy
import spacy

nlp = spacy.load('en_core_web_sm') 
```

```python
# create a function count number of NE tags for a specific NER category 
def ne_spacy_tagger(row, col_name, ner_cat):
  tag_list = []
  text = row[col_name]
  doc = nlp(text)
  for ent in doc.ents:
    entity = ent.text
    en_category = ent.label_
    t = (en_category, entity)
    if en_category == ner_cat:
      tag_list.append(t)
  return len(set(tag_list))
```

```python
# Need to drop rows that have an empty review
df_gb.dropna(inplace = True)
df_gb['Review'].isnull().values.any()
```

```python
df_ner_spacy = df_gb.copy()

# create NER category Organization feature

%time df_ner_spacy['NER_ORG'] = df_ner_spacy.apply(lambda row:ne_spacy_tagger(row, 'Review', 'ORG'), axis =1)

%time df_ner_spacy['NER_DATE'] = df_ner_spacy.apply(lambda row:ne_spacy_tagger(row, 'Review', 'DATE'), axis =1)

%time df_ner_spacy['NER_Person'] = df_ner_spacy.apply(lambda row:ne_spacy_tagger(row, 'Review', 'PERSON'), axis =1)
```

```python
# Contraction expansion
!pip install contractions
import contractions
```

```python
# Define contraction funtion
def contraction_expansion(text):
  expanded_word = []    
  for word in text.split():
  # using contractions.fix to expand
    expanded_word.append(contractions.fix(word))   
   
  return  ' '.join(expanded_word)
```

```python
# Test to see if there was any empty reviews in the dataset
df_gp['Review'].isnull().values.any()
```

```python
# import POS libiaries
import nltk
nltk.download('punkt') 
nltk.download('averaged_perceptron_tagger') 
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer
```

```python
# create POS tag
%time df_gp['Review_tagged'] = df_gp['Review_Token'].apply(nltk.pos_tag)
df_gp.head(3)
```

```python
# create a function to lemmatize words based on different POS tag types; 
# slightly change from earlier as input is not a tagged corpus, but an individual tagged column
def lemmatize_POS (tagged_columns):
  wnl = WordNetLemmatizer()
  text=[]
  for word, tag in tagged_columns:
    if tag.startswith('NN'):
      token = wnl.lemmatize(word, pos='n') # n is NOUN 
    elif tag.startswith('VB'):
      token = wnl.lemmatize(word, pos='v') # v is VERB
    elif tag.startswith('JJ'):
      token = wnl.lemmatize(word, pos='a') # a is ADJ
    elif tag.startswith('RB'):
      token = wnl.lemmatize(word, pos='r') # r is ADV
    else:
      token = word
    text.append(token)
  #cleaned_corpus.append(text)
  return text
```

```python
%time df_gp['Review_lemmas'] = df_gp['Review_tagged'].apply(lemmatize_POS)
df_gp['Review_lemmas'].head(3)
```

```python
# Define preprocessing function
from nltk.corpus import stopwords
nltk.download('stopwords')
from string import punctuation

def preprocessing(tokens):
  tokens = [token.lower() for token in tokens] # lowercasing
  
  tokens = [token for token in tokens if not token.isdigit()] # remove digit

  tokens = [token for token in tokens if token not in punctuation] # remove punctuations

  mystopwords = set(stopwords.words("english")) 
  tokens = [token for token in tokens if token not in mystopwords] # remove stopwords
  
  tokens = [word for word in tokens if len(word)>=3] # remove words with one or two characters

  return tokens
```

```python
# now let's apply text normalization
%time df_gp['Review_lemmas_final'] = df_gp['Review_lemmas'].apply(preprocessing)
df_gp['Review_lemmas_final'].head(3)
```

```python
# count the frequency of each unique token
from nltk import FreqDist
# use FreqDis() to find the frequency disctibution for each token
%time fdist_reviews= FreqDist([token for sublist in df_gp['Review_lemmas_final'].values for token in sublist])
print('The total number of tokens:',fdist_reviews.N())
print('The number of unique tokens:', len(fdist_reviews))
```

```python
# let's sort the items based on the frequency count
fdist_reviews_list = sorted(fdist_reviews.items(), key=lambda x:x[1], reverse= True)
type(fdist_reviews_list) # this is a list of tuples
```

```python
# create a text column with tagged tokens
%time df_gp['Cleaned_review_lemma'] = df_gp['Review_lemmas_final'].apply(lambda row: ' '.join(str(x) for x in row))
df_gp.head()
```

## Question 7: BOW vectorization


```python
# create a list of top 200 tokens
token_list = [token[0] for token in fdist_reviews_list[:200]]
```

```python
print(len(token_list))
print(token_list) # take a look at the token features
```

```python
# using token frequency as features
def frequency_feature(row, col_name, word):
  if word in row[col_name]:
    return row[col_name].count(word)
  else:
    return 0
```

```python
# Frequency-based BOW
df_freq = df_gp.copy()

import time 
start = time.time()

# BOW as feature names
for word in token_list:
  df_freq[word] = df_freq.apply(lambda row:frequency_feature(row, 'Cleaned_review_lemma', word), axis =1)

print((time.time() - start))
```
