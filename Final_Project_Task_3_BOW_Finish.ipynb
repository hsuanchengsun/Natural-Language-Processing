### BOW as features

Keep Top Frequent Tokens

```python
# you only need to run pip install once
#! pip install nltk==3.6.2 
# you will need to import packages each time your runtime restart
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, wordpunct_tokenize
nltk.download('punkt') 
nltk.download('averaged_perceptron_tagger') 
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer
```

```python
# tokenize each review in the dataframe
%time df_aggre['Review_Token'] = df_aggre['cleaned_review'].apply(wordpunct_tokenize)
df_aggre.head(3)
```

```python
# count the frequency of each unique token
from nltk import FreqDist
# use FreqDis() to find the frequency disctibution for each token
%time fdist_reviews= FreqDist([token for sublist in df_aggre['Review_Token'].values for token in sublist])
```

```python
print('The total number of tokens:',fdist_reviews.N())
print('The number of unique tokens:', len(df_aggre))
```

```python
# let's sort the items based on the frequency count
fdist_reviews_list = sorted(fdist_reviews.items(), key=lambda x:x[1], reverse= True)
type(fdist_reviews_list) # this is a list of tuples
```

```python
# using token frequency as features
def frequency_feature(row, col_name, word):
  if word in row[col_name]:
    return row[col_name].count(word)
  else:
    return 0
```

```python
# fdist_review_list is a list of tuples, so simply print them will show the token and its count
# see the top 10 tokens and its count
for x in fdist_reviews_list[:20]:
    print (x[0], x[1])
```

```python
# create a list of top 500 tokens
token_list = [token[0] for token in fdist_reviews_list[:500]]
```

```python
print(len(token_list))
print(token_list) # take a look at the token features
```

```python
import time 
start = time.time()

# BOW as feature names
for word in token_list:
  df_aggre[word] = df_aggre.apply(lambda row:frequency_feature(row, 'review_lemmas', word), axis =1)

print((time.time() - start))
```
